Excluir linhas da tabela
- .drop()

```python
vendas_df = pd.read_excel('vendas.xlsx')
display(vendas_df)

#Excluir coluna
#vendas_df = vendas_df.drop('data', axis=1)
#display(vendas_df)

#Excluir linha
vendas_df = vendas_df.drop(4, axis=0)
display(vendas_df)
```

Tratamento de linhas e colunas vazias.

- .dropna para tratar linhas vazias
- .fillna para tratar linhas vazias
- .fillna para fazer metricas em linhas vazias 

```python
vendas_df = pd.read_excel('vendas.xlsx')
display(vendas_df)

#vendas_df = vendas_df.dropna()
#display(vendas_df)

#vendas_df['preco'] = vendas_df['preco'].fillna(15)
#display(vendas_df)

vendas_df['preco'] = vendas_df['preco'].fillna(vendas_df['preco'].mean())
display(vendas_df)
```

Ajustar tabelas vazias utilizando o parâmetro .ffill()

```python
vendas_df = pd.read_excel('vendas.xlsx')
display(vendas_df)

vendas_df = vendas_df.ffill()
display(vendas_df)
```

Contar valores dentro de uma tabela
 - .value_counts()
 - .groupby()

```python
contador = vendas_df['preco'].value_counts()
display(contador)
```

```python
fatoramento_produtos = vendas_df[['produto','preco']].groupby('produto').sum()

display(vendas_df)
display(fatoramento_produtos)
```

Mesclagem de tabelas
 - Merge utilizado para mesclar tabelas

```python
gerentes_df = pd.read_excel('gerentes.xlsx')

display(vendas_df)

gerentes_lojas = vendas_df.merge(gerentes_df)
display(gerentes_lojas)
```
